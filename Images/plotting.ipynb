```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H0 = 68
omegas = np.arange(0.01, 0.99, 0.01)
```

```python
def t0_q54(Omega, H0):
    return 1/H0 * (1/(1-Omega) - Omega/(2*(1-Omega)**(3/2))* np.arccosh((2-Omega)/Omega))
```

```python
plt.plot(omegas, t0_q54(omegas, H0))
plt.title("t0 vs Omega0")
plt.xlabel("Omega0")
plt.ylabel("t0 (s * MPc/km)")
plt.savefig("Q5-4.pdf")
plt.show()
```

```python
def t0_q53(omegas, H0):
    return 1/H0 * 0.5 * omegas/(omegas - 1)**(3/2) * np.arccos((2 - omegas)/omegas) - 1/(omegas-1)
```

```python
omegas = np.arange(1.01, 3, 0.01)
```

```python
plt.plot(omegas, t0_q53(omegas, H0))
plt.title("t0 vs Omega0")
plt.xlabel("Omega0")
plt.ylabel("t0 (s * MPc/km)")
plt.savefig("Q5-3.pdf")
plt.show()
```

```python
Ts = np.arange(3000, 4501, 1)
```

```python
eta = 4E-10
```

```python
def X(S):
    return (-1 + np.sqrt(1 + 4 * S))/(2 * S)

def S(eta, T):
    Q = 13.6
    k = 8.62E-5
    mec2 = 511000
    return 3.84 * eta * (k * T/mec2)**(3/2) * np.exp(Q/k/T)
```

```python
plt.plot(Ts, X(S(eta, Ts)))
plt.title("X vs. T for eta = 4E-10")
plt.xlabel("T (K)")
plt.ylabel("X")
plt.savefig("Q8-1eta4.pdf")
plt.show()
```

```python
eta = 8E-10
```

```python
plt.plot(Ts, X(S(eta, Ts)))
plt.title("X vs. T for eta = 8E-10")
plt.xlabel("T (K)")
plt.ylabel("X")
plt.savefig("Q8-1eta8.pdf")
plt.show()
```

```python
thetas = np.linspace(0, 2*np.pi, 1000)
```

```python
def a(theta, Omega0):
    return 0.5 * Omega0/(Omega0 - 1) * (1 - np.cos(theta))

def t(theta, Omega0, H0):
    return 0.5 / H0 * Omega0/(Omega0-1)**(3/2) * (theta - np.sin(theta))
```

```python
plt.plot(t(thetas, 2, 1), a(thetas, 2))
plt.title(r"$t(\theta)$ vs. $a(\theta)$")
plt.xlabel(r"$a(\theta)$")
plt.ylabel(r"$t(\theta)$")
plt.savefig("Q5-A.pdf")
plt.show()
```
